## 일자별 거래실적 추이 (거래대금)

```python
#지정된 기간 동안의 일자별로 정렬된 (순매수) 거래량
df = stock.get_market_trading_value_by_date("20210101", "20210701", "005930")
print(df.head())
```

```python
# 매수 혹은 매도 금액을 조회하고 싶다면 on 파라미터에 입력할 수 있습니다. 입력하지 않는다면 순매수 금액이 조회
df = stock.get_market_trading_value_by_date("20210115", "20210122", "005930", on='매도')
print(df.head())
```

```python
# detail 옵션을 추가하면 상세한 정보를 얻을 수 있습니다. 옵션은 KOSPI, KOSDAQ, KONEX 조회할 때만 유효

df = stock.get_market_trading_value_by_date("20210115", "20210122", "KOSPI", etf=True, etn=True, elw=True, detail=True)
print(df.head())
```
